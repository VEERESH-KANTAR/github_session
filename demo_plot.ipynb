```python
import numpy as np
import matplotlib.pyplot as plt

# X values
x = np.arange(0, 11, 1)

# y = x
y = x

# Plot
plt.plot(x, y, marker="o", label="y = x")

# Labels and styling
plt.title("Line Plot Demo: y = x")
plt.xlabel("X axis")
plt.ylabel("Y values")
plt.grid(True)
plt.legend()

plt.show()
```
